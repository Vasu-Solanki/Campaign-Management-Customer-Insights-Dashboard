```python
# importing common libraries for data analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# to see all the columns
pd.set_option("display.max_columns", None)
```

```python
# importind market campaign data
df = pd.read_csv("/content/marketing_campaign_22-24.csv")
df.head()
```

# 1. Basic Exploration & fixing

```python
# Shape of our dataset (rows , columns)
df.shape
```

```python
# Here we checking data types are in correct format or not.
df.info()
```

```python
# Here in Dt_Customer dtype from obj -> Dateformat
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
```

```python
# Dtype change successfully
df[["Dt_Customer"]].info()
```

```python
# There are no duplicate values
df.duplicated().sum()
```

```python
# Checking distribution or any mis spell values in object columns
for i in df.select_dtypes(include = "object"):
  print(i, df[i].nunique())
  print(df[i].value_counts())
  print()
```

```python
# Drop rows with 'Alone', 'YOLO', or 'Absurd' in 'Marital_Status'
df = df[~df['Marital_Status'].isin(['Alone', 'YOLO', 'Absurd'])]
print(df['Marital_Status'].value_counts())
```

```python
df.describe(percentiles=[0.01,0.99 ]).T
```

```python
# Checking null values in columns
df.isna().sum()
# Inocome colums have 24 null values
```

```python
# displaying rows where income have null values
df[df['Income'].isnull()].head()
```

```python
# Fill null values in 'Income' with the mean income grouped by 'Marital_Status', and 'Education'.
df["Income"] = df.groupby([ 'Marital_Status', 'Education'])["Income"].transform(lambda x : x.fillna(x.median()))
```

```python
# Clipping outliers using a for loop
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

for col in numerical_cols:
    lower_bound = df[col].quantile(0.01)
    upper_bound = df[col].quantile(0.99)
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

display(df.describe(percentiles=[0.01, 0.99]).T)
```

# 2. Feature Engineering

```python
# Extracting year, month with name, month quarter,days and phase of month from "Dt_customer" (joining data of the customer)
df["Cust_enroll_year"] = df["Dt_Customer"].dt.year
df["Cust_enroll_month"] = df["Dt_Customer"].dt.month_name()
df["Cust_enroll_Quarter"] = df["Dt_Customer"].dt.quarter.map({1:"Q1", 2:"Q2", 3:"Q3", 4:"Q4"})
df["Cust_enroll_day"] = df["Dt_Customer"].dt.day_name()
df["Cust_enroll_month_Phase"] = df["Dt_Customer"].apply(lambda x : "Start" if 1 <= x.day <= 10 else ( "Middle" if 11 <= x.day <= 20 else "End"))


# Extracting age and joining age from df
from datetime import datetime
df["Age"] = 2024 - df["Year_Birth"]
df["Joining_Age"] = df["Cust_enroll_year"] - df["Year_Birth"]
```
